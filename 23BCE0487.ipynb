```python
X = df.iloc[:, [3, 5]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 2)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of cars')
plt.xlabel('Horse Power')
plt.ylabel('acceleration')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = df.select_dtypes(include=[np.number])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_cols)
```

```python
pca = PCA()
pca_result = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
print("Eigenvalues:", pca.explained_variance_)
print("Explained Variance Ratio:", explained_variance)
print("Cumulative Explained Variance:", cumulative_variance)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--', label='Individual')
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='s', linestyle='-', label='Cumulative')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.legend()
plt.show()
```

```python
n_comp = 2  # Adjust based on scree plot
pca = PCA(n_components=n_comp)
pca_result = pca.fit_transform(scaled_data)
```

```python
# Plot PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=df['cylinders'])  # Adjust hue if needed
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Components Plot")
plt.show()
```

```python
# Summary of PCA
print(pd.DataFrame(pca.components_, columns=numeric_cols.columns, index=[f'PC{i+1}' for i in range(n_comp)]))
```
