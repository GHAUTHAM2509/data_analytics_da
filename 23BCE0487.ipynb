```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Auto.csv")
df.head(10)
```

```python
df["fuelEfficiency"] = df["mpg"] / df["cylinders"]
df.head(10)
```

```python
# example of binning
mpg_bin = pd.cut(df['mpg'],bins = 5)
mpg_bin.value_counts()
```

```python
#outlier detection in weight
weight = df['weight'].mean() + 2*df['weight'].std()
weight
```

```python
# First 30 largest mpg and draw a bar chart
df.mpg.value_counts().nlargest(30).plot(kind='bar', figsize=(10,6))
plt.title("Number of cars by mpg")
plt.ylabel('Number of cars')
plt.xlabel('Mileage per gallon')
```

```python
df.acceleration.value_counts().nlargest(30).plot(kind='bar', figsize=(10,6))
plt.title("Number of cars by acceleration")
plt.ylabel('Number of cars')
plt.xlabel('Acceleration')
```

```python
# Distribution plot
sns.FacetGrid(df).map(sns.distplot,"weight").add_legend()
```

```python
#boxplot for no of cylinders of cars
sns.boxplot(x="mpg",data=df)
plt.show()
```

```python
#boxplot for no of cylinders of cars
sns.boxplot(x="cylinders",data=df)
plt.show()
```
